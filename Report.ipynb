# Preparation

### The dataset choice  
I selected SportsStats data, as its size is smaller than the others. This will ease data wrangling and analysis with my laptop.

### Initial exploration of the data

Out of 271116 data points, 9474 lack age value, 60171 lack height value, 62875 lack weight value. Sex values are complete. Games data and its attributes (year, city, etc.) are complete.  
  
Team names can change. Using NOC instead would prove more consistent.  

# Project Proposal
In this project, I am going to analyze the summer and winter Olympic games data. I am going to attempt to figure out important changes in terms of diversity and performance. I am also going to try to find correlation between several factors (such as representing the host country) and team performance.
Anyone interested in understanding the demographics of Olympic games could use this analysis.

### Questions  
Q1: How has the male:female ratio changed through time?  
Q2: Does country performance by year vary more in Winter Olympics?  
Q3: Does performance of a country in winter olympics correlate with its performance in summer olympics?  

### Hypotheses  
H1: Decreased.   
H2: Yes.  
H3: Yes.  
  
### Approach  
I am going to draw a simple histogram for the first question and I am going to use A/B testing on the other two of my hypotheses.

# Descriptive Analysis


In the last century, in both competition, medal ratio has fluctuated widely, but ended up stabilized. This might be explained establishment of norms in these matters.


Relative percentages of gold, silver, and bronze medals has stabilized as well, probably due to the above-mentioned reason.


Female to male ratio has indeed increased through time. There is, however, an interesting detail: around the WW2, ratio in Summer Olympics has decreased steeply, only to regain momentum later. I am unable to explain this phenomenon without further analysis.
